```python
import pandas as pd
import numpy as np
df = pd.read_csv('car_dataset.data')
df.head()
```

```python
headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", type(headers_data)) 
```

```python
df = pd.read_csv('car_dataset.data', names = headers_data)
df.head(2)      
```

Types of missing values

1. Standard missing values: Empty (pandas can interpret it on its own.
2. Non standard: N/a, n/a, --, ? etc values entered manually which are not understood by system and is either entred by user or any filter. Pandas cant identify it.
3. Unexpected missing value: Lets say you have to enter y or n but you enter 0 or 1 or somthing else which shouldnot be there.

```python
# Replacing non standard values using replace so that pandas can understand it

df.replace('?', np.nan, inplace = True)
df
```

```python
# using a loop to get count of null values
missing_data = df.isnull()
for column in headers_data :
    print(column)
    print(missing_data[column].value_counts()) # gives unique value and its count in the data
    print(' ')
```

Methods to deal with missing values
1. Drop the row/column
2. Replace missing values (Avg value/ median value(incase of outliers)) oe mode (most frequency)
3. Leave it as it is

Stratey to be used in this particular dataset
1. By mean: Normalised losses, bore, stroke, hp, peakrpm
2. Mode:  No. of doors
3. Remove: Price ( SInce its a tageted variable, any changes made to it will affect outcome)

```python
# converting datatype the columns
av_norm_loss = df['normalized-losses'].astype('float').mean()
av_norm_loss
```

```python
df['normalized-losses'].replace(np.nan, av_norm_loss, inplace = True)
df.head(6)
```

```python
# converting datatype the columns
av_bore = df['bore'].astype('float').mean()
df['bore'].replace(np.nan, av_bore, inplace = True)

av_stroke = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, av_stroke, inplace =True)

av_hp = df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, av_hp, inplace = True)

av_rpm = df['peak-rpm'].astype('float').mean()
df['peak-rpm'].replace(np.nan, av_rpm, inplace = True)

df.head(10)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['body-style'].value_counts()
```

```python
# Since most of the cars are4 doors then we can replace by most frequrnt values
df['num-of-doors'].replace(np.nan, 'four', inplace= True)
```

```python
# Dropping columns/Rows
df.dropna(subset = ['price'],inplace =True ,  axis = 0)
```

```python
# resetting index
df.reset_index(drop = True, inplace = True)
```

```python
# Now checking data in all columns 

missing_values = df.isnull()

for columns in headers_data:
    print(columns)
    print(missing_values[columns].value_counts())
    print(' ')
```

### Using scikit learn


```python
dff = pd.read_csv('car_dataset.data', names = headers_data )
dff.head()
```

```python
from sklearn.impute import SimpleImputer as SI
```

```python
x = SI(missing_values=np.nan, strategy='mean')
```

```python
dff[['bore','stroke','horsepower','peak-rpm']] = x.fit_transform(dff[['bore','stroke','horsepower','peak-rpm']]).astype('object')
dff.head()
```

```python
dff['bore'].unique()
```

```python
dff.replace('?', np.nan, inplace = True)
dff
```

```python
dff[['normalized-losses','bore','stroke','horsepower','peak-rpm']] = x.fit_transform(dff[['normalized-losses','bore','stroke','horsepower','peak-rpm']]).astype('object')
dff.head()
```

```python
# Now using imputer to replace it with most frequent value\

median = SI(missing_values=np.nan, strategy='most_frequent')
dff[['num-of-doors']] = median.fit_transform(dff[['num-of-doors']])
```

```python
dff.info()
```

```python
dff.dropna(subset= ['price'],axis=0, inplace=True)
dff.reset_index(drop = True, inplace = True)
dff.info()
```

```python
missing_val = dff.isnull()

for col in headers_data:
    print(col)
    print(missing_val[col].value_counts())
    print(' ')
```
