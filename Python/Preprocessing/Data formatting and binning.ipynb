### Method in which we check that our data is in correct or desired format. (Correct data type)

```python
import numpy as np
import pandas as pd
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv('car_dataset.data', names = headers )
```

```python
df.head()
```

```python
df.replace('?', np.nan, inplace = True)
```

In case of replacing values with pandas, it was clear that our final data type was not changed from object to floar however in case of scikit learn data type was chnaged to float as its an advanced library.
Now changing data type of each column.
But if we export the data and then reload it, it will be in desired state.

```python
# First manually repalcing the missing values in our dataset

loss_mean = df['normalized-losses'].astype('float').mean()
loss_mean
```

```python
df['normalized-losses'].replace(np.nan, loss_mean, inplace= True)
df.head()
```

```python
# converting datatype the columns
av_bore = df['bore'].astype('float').mean()
df['bore'].replace(np.nan, av_bore, inplace = True)

av_stroke = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, av_stroke, inplace =True)

av_hp = df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, av_hp, inplace = True)

av_rpm = df['peak-rpm'].astype('float').mean()
df['peak-rpm'].replace(np.nan, av_rpm, inplace = True)

df.head(10)
```

```python
df.info()
```

```python
df[['bore','stroke', 'normalized-losses','horsepower','peak-rpm','price']] = df[['bore','stroke', 'normalized-losses','horsepower','peak-rpm','price']].astype(float)
df.info()
```

Data binning: Incase of data, we design a range of that value set. Example low, med, high prices based on our filter. Here low,med ,high are the bins crreated and this process is binning.

```python
# Binning on horsepower column
#convertingit intp integer
df['horsepower'] = df['horsepower'].astype(int)
```

Visualizing the column as a Histogram

```python
import matplotlib.pyplot as plt

plt.hist(df['horsepower'])
plt.xlabel('Horse power')
plt.ylabel('Frequency')
```

```python
#Np.linspace divides the first and last number in equidiant gps as specified
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
#Creating bins names
bin_names =['Low', 'Med', 'High']
```

```python
df['horsepower_binned'] = pd.cut(df['horsepower'], bins, labels= bin_names, include_lowest= True)
df[['horsepower','horsepower_binned']]
```

```python
df['horsepower_binned'].value_counts()
```

```python
plt.bar(bin_names, df['horsepower_binned'].value_counts())
```
